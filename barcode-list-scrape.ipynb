```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


cookies = {
    'PHPSESSID': 's17np6nj2ttma93rb65qkg2bva',
    '_ga': 'GA1.2.2018648644.1686816680',
    '_gid': 'GA1.2.240899695.1686816680',
    '_ga_SN0NNFD7XH': 'GS1.2.1686816680.1.1.1686818041.0.0.0',
}

headers = {
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
    'Accept-Language': 'en-US,en;q=0.9',
    'Cache-Control': 'no-cache',
    'Connection': 'keep-alive',
    # 'Cookie': 'PHPSESSID=s17np6nj2ttma93rb65qkg2bva; _ga=GA1.2.2018648644.1686816680; _gid=GA1.2.240899695.1686816680; _ga_SN0NNFD7XH=GS1.2.1686816680.1.1.1686818041.0.0.0',
    'Pragma': 'no-cache',
    'Sec-Fetch-Dest': 'document',
    'Sec-Fetch-Mode': 'navigate',
    'Sec-Fetch-Site': 'none',
    'Sec-Fetch-User': '?1',
    'Upgrade-Insecure-Requests': '1',
    'User-Agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Mobile Safari/537.36',
    'sec-ch-ua': '"Not.A/Brand";v="8", "Chromium";v="114", "Google Chrome";v="114"',
    'sec-ch-ua-mobile': '?1',
    'sec-ch-ua-platform': '"Android"',
}

def parse_page(page_content):
    # Create a BeautifulSoup object and specify the parser
    soup = BeautifulSoup(page_content, 'html.parser')

    raw_table = str(soup.find('table', attrs={'class': 'randomBarcodes'}))
    products = pd.read_html(raw_table)
    return products[0]

def scrape_webpage(search_request):
    # Send a GET request
    params = {
        'barcode': search_request,
    }
    response = requests.get(
        'https://barcode-list.ru/barcode/RU/%D0%9F%D0%BE%D0%B8%D1%81%D0%BA.htm',
        params=params,
        cookies=cookies,
        headers=headers,
    )

    # If the GET request is successful, the status code will be 200
    if response.status_code == 200:
        return parse_page(response.content)
    else:
        print(response.status_code)

# Call the function and print the result
products = scrape_webpage('липтон')
```
